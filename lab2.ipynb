```python
from pypdf import PdfReader 
import glob
import os
import pandas as pd
import re
import numpy as np 
import tabula
```

```python
def find_str(pattern, text):
    match = re.search(pattern, text)
    if match:
        val = match.group(1)
        return val
    else:
        return None
```

```python
cwd = os.getcwd()
directory = cwd + '/datasets/old_data/lab/'
pdf_files = glob.glob(os.path.join(directory, '**', '*.pdf'), recursive=True)
```

```python
columns = ['patient', 'ldl', 'vldl', 'gfr', 'bilirubin_in', 'aks', 'bun', 'crea', 
           'tri', 'colestrol', 'hdlk', 'sgpt', 'sgot', 'amilaz', 'bilirubin_tot',
           'bilirubin_dir', 'cal', 'sod', 'pot', 'clor', 'hemoliz', 'ikter', 'lipemi', 'wbc',
           'rbc', 'hgb', 'hct', 'plt', 'mcv', 'mch', 'mchc', 'rdwcv', 'rdwsd', 'mpv', 'pct',
           'pdw', 'ne', 'lym', 'mono', 'eos', 'baso', 'neu', 'lym_hatch', 'mono_hatch',
           'eos_hatch', 'baso_hatch', 't3', 't4', 'tsh', 'folat', 'b12','TROPONIN I', 'CK-MB', 
           'BNP', 'HbA1C']


df = pd.DataFrame(columns=columns)
```

```python
for pdf in pdf_files:
    reader = PdfReader(pdf)
    all_text = ""
    for page in reader.pages:
        all_text += page.extract_text() + "\n"
    pattern = r'LDL K [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dL'
    ldl = find_str(pattern, all_text)

    pattern = r'VLDL K [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dl'
    vldl = find_str(pattern, all_text)

    pattern = r'GFR \(MDRD\) [\[A-Z\] ]*(\d+[,\d][\d]*)ml/dk'
    gfr = find_str(pattern, all_text)

    pattern = r'BİLİRÜBİN-İNDİREKT [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dl'
    bilirubin_in = find_str(pattern, all_text)

    pattern = r'AKŞ GLUKOZ [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dL'
    aks = find_str(pattern, all_text)

    pattern = r'BUN [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dl'
    bun = find_str(pattern, all_text)

    pattern = r'KREATİNİN \(CREA\) [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dl'
    crea = find_str(pattern, all_text)

    pattern = r'TRİGLİSERİD [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dl'
    tri = find_str(pattern, all_text)

    pattern = r'T.KOLESTEROL [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dL'
    colestrol = find_str(pattern, all_text)

    pattern = r'HDL-K [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dl'
    hdlk = find_str(pattern, all_text)

    pattern = r'SGPT \(ALT\) [\[A-Z\] ]*(\d+[,\d][\d]*)U/L'
    sgpt = find_str(pattern, all_text)

    pattern = r'SGOT\(AST\) [\[A-Z\] ]*(\d+[,\d][\d]*)U/L'
    sgot = find_str(pattern, all_text)

    pattern = r'AMİLAZ [\[A-Z\] ]*(\d+[,\d][\d]*)U/L'
    amilaz = find_str(pattern, all_text)

    pattern = r'BİLİRÜBİN-TOTAL [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dL'
    bilirubin_tot = find_str(pattern, all_text)

    pattern = r'BİLİRÜBİN-DİREKT [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dl'
    bilirubin_dir = find_str(pattern, all_text)
    
    pattern = r'KALSİYUM \(Ca\) [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dl'
    cal = find_str(pattern, all_text)

    pattern = r'SODYUM \(Na\) [\[A-Z\] ]*(\d+[,\d][\d]*)mmol/L'
    sod = find_str(pattern, all_text)

    pattern = r'POTASYUM \(K\) [\[A-Z\] ]*(\d+[,\d][\d]*)mmol/L'
    pot = find_str(pattern, all_text)

    pattern = r'KLOR \(Cl\) [\[A-Z\] ]*(\d+[,\d][\d]*)mmol/L'
    clor = find_str(pattern, all_text)

    pattern = r'Serum İndeksi \(Hemoliz\) [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dL'
    hemoliz = find_str(pattern, all_text)

    pattern = r'Serum İndeksi \(İkter\) [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dL'
    ikter = find_str(pattern, all_text)

    pattern = r'Serum İndeksi \(Lipemi\) [\[A-Z\] ]*(\d+[,\d][\d]*)mg/dL'
    lipemi = find_str(pattern, all_text)

    pattern = r'WBC [\[A-Z\] ]*(\d+[,\d][\d]*)10\^3/mL'
    wbc = find_str(pattern, all_text)

    pattern = r'RBC [\[A-Z\] ]*(\d+[,\d][\d]*)10\^6/mL'
    rbc = find_str(pattern, all_text)

    pattern = r'HGB [\[A-Z\] ]*(\d+[,\d][\d]*)g/dL'
    hgb = find_str(pattern, all_text)

    pattern = r'HCT [\[A-Z\] ]*(\d+[,\d][\d]*)\%'
    hct = find_str(pattern, all_text)

    pattern = r'PLT [\[A-Z\] ]*(\d+[,\d][\d]*)10\*3/µL'
    plt = find_str(pattern, all_text)

    pattern = r'MCV [\[A-Z\] ]*(\d+[,\d][\d]*)fL'
    mcv = find_str(pattern, all_text)

    pattern = r'MCH [\[A-Z\] ]*(\d+[,\d][\d]*)pg'
    mch = find_str(pattern, all_text)

    pattern = r'MCHC [\[A-Z\] ]*(\d+[,\d][\d]*)g/dL'
    mchc = find_str(pattern, all_text)

    pattern = r'RDW-CV [\[A-Z\] ]*(\d+[,\d][\d]*)\%'
    rdwcv = find_str(pattern, all_text)

    pattern = r'RDW-SD [\[A-Z\] ]*(\d+[,\d][\d]*)fL'
    rdwsd = find_str(pattern, all_text)

    pattern = r'MPV [\[A-Z\] ]*(\d+[,\d][\d]*)fL'
    mpv = find_str(pattern, all_text)

    pattern = r'PCT [\[A-Z\] ]*(\d+[,\d][\d]*)\%'
    pct = find_str(pattern, all_text)

    pattern = r'PDW [\[A-Z\] ]*(\d+[,\d][\d]*)fL'
    pdw = find_str(pattern, all_text)

    pattern = r'NE% [\[A-Z\] ]*(\d+[,\d][\d]*)\%'
    ne = find_str(pattern, all_text)

    pattern = r'LYM% [\[A-Z\] ]*(\d+[,\d][\d]*)\%'
    lym = find_str(pattern, all_text)

    pattern = r'MONO% [\[A-Z\] ]*(\d+[,\d][\d]*)\%'
    mono = find_str(pattern, all_text)

    pattern = r'EOS% [\[A-Z\] ]*(\d+[,\d][\d]*)\%'
    eos = find_str(pattern, all_text)

    pattern = r'BASO% [\[A-Z\] ]*(\d+[,\d][\d]*)\%'
    baso = find_str(pattern, all_text)

    pattern = r'NEU# [\[A-Z\] ]*(\d+[,\d][\d]*)10\^3/mL'
    neu = find_str(pattern, all_text)

    pattern = r'LYM# [\[A-Z\] ]*(\d+[,\d][\d]*)10\^3/mL'
    lym_hatch = find_str(pattern, all_text)

    pattern = r'MONO# [\[A-Z\] ]*(\d+[,\d][\d]*)10\^3/mL'
    mono_hatch = find_str(pattern, all_text)

    pattern = r'EOS# [\[A-Z\] ]*(\d+[,\d][\d]*)10\^3/mL'
    eos_hatch = find_str(pattern, all_text)

    pattern = r'BASO# [\[A-Z\] ]*(\d+[,\d][\d]*)10\^3/mL'
    baso_hatch = find_str(pattern, all_text)

    pattern = r'Free T3 [\[A-Z\] ]*(\d+[,\d][\d]*)pg/mL'
    t3 = find_str(pattern, all_text)

    pattern = r'Free T4 [\[A-Z\] ]*(\d+[,\d][\d]*)ng/dL'
    t4 = find_str(pattern, all_text)

    pattern = r'TSH [\[A-Z\] ]*(\d+[,\d][\d]*)uIU/mL'
    tsh = find_str(pattern, all_text)

    pattern = r'FOLAT [\[A-Z\] ]*(\d+[,\d][\d]*)ng/mL'
    folat = find_str(pattern, all_text)

    pattern = r'ViİTAMİN B12 [\[A-Z\] ]*(\d+[,\d][\d]*)pg/mL'
    b12 = find_str(pattern, all_text)

    pattern = r'TROPONİN I [\[A-Z\] ]*(\d+[,\d][\d]*)ng/L'
    troponin = find_str(pattern, all_text)
    pattern = r'CK-MB [\[A-Z\] ]*(\d+[,\d][\d]*)ng/mL'
    ckmb = find_str(pattern, all_text)
    pattern = r'BNP [\[A-Z\] ]*(\d+[,\d][\d]*)pg/mL'
    bnp = find_str(pattern, all_text)
    pattern = r'HbA1C \(mmol/mol\) [\[A-Z\] ]*(\d+[,\d][\d]*)mmol/L'
    hb1acmol = find_str(pattern, all_text)
    patient = int(re.search(r'/(\d+)\.pdf$', pdf).group(1))
    df.loc[len(df)] = [patient, ldl, vldl, gfr, bilirubin_in, aks, bun, crea, tri, colestrol,
                       hdlk, sgpt, sgot, amilaz, bilirubin_tot, bilirubin_dir, cal, sod, pot,
                       clor, hemoliz, ikter, lipemi ,wbc, rbc, hgb, hct, plt, mcv, mch, mchc, 
                       rdwcv, rdwsd, mpv, pct, pdw, ne, lym, mono, eos, baso, neu, lym_hatch, 
                       mono_hatch, eos_hatch, baso_hatch, t3, t4, tsh, folat, b12, troponin, 
                       ckmb, bnp, hb1acmol]

```
