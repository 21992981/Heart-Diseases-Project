```python
import pandas as pd
import os
```

```python
cwd = os.getcwd()
timeseries_df = pd.read_csv(cwd + "/datasets/timeseriesdata.csv")
timeseries_df = timeseries_df.drop('Unnamed: 0', axis=1)
timeseries_df
```

```python
wav_df = pd.read_csv(cwd + "/datasets/wav_encoding.csv")
wav_df = wav_df.drop('Unnamed: 0', axis=1)
wav_df
```

```python
timeseries_df['rate_avg'] = wav_df['rate_avg']
timeseries_df['rate_std'] = wav_df['rate_std']
timeseries_df['rate_min'] = wav_df['rate_min']
timeseries_df['rate_max'] = wav_df['rate_max']
```

```python
timeseries_df.to_csv('datasets/timeseries_with_wav.csv')
```
