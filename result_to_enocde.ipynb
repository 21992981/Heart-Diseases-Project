```python
def apply_dummies(group):
    dummies = pd.get_dummies(group['result'], prefix='result')
    group = pd.concat([group, dummies], axis=1)
    group = group.drop('result', axis=1)
    return group

# Apply the function to each group and combine the results
df_grouped = df.groupby('patient').apply(apply_dummies).reset_index(drop=True)

# Display the resulting DataFrame
df_grouped
```
